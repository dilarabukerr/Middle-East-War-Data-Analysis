```python
#DATA PREPROCESSING

#yaş sütununu ortalama değerle değiştir.
age_mean = df['age'].mean()
df['age'].fillna(age_mean, inplace = True ) #fillna() eksik değerleri dolduruyor.
```

```python
#gender, place_of_residence, place_of_residence_district, type_of_injury, notes sütunlarını, mod değerleriyle değiştir.

gender_mode = df['gender'].mode()[0]
place_of_residence_mode = df['place_of_residence'].mode()[0]
place_of_residence_district_mode = df['place_of_residence_district'].mode()[0]
type_of_injury_mode = df['type_of_injury'].mode()[0]
notes_mode = df['notes'].mode()[0]

#eksik deperleri mode değerleri ile doldur.
df['gender'].fillna(gender_mode, inplace=True)
df['place_of_residence'].fillna(place_of_residence_mode, inplace=True)
df['place_of_residence_district'].fillna(place_of_residence_district_mode, inplace=True)
df['type_of_injury'].fillna(type_of_injury_mode, inplace=True)
df['notes'].fillna(notes_mode, inplace=True)
```

```python
#took_part_in_the_hostilities, ammunition sütunlarını kaldırma
df.drop(['took_part_in_the_hostilities', 'ammunition'],  axis=1, inplace=True)
```

```python
#eksik değerleri tekrar kontrol et..
df.isna().sum()
```

```python
df['citizenship'].value_counts() #citizenship sütunundaki benzersiz değerleri sayar. (hangi 'citizenship'ten kaç tane var?)
```

```python
#Jordanian ve American olan satırları silelim..(filtreliyoruz)
df = df[~df['citizenship'].isin(['Jordanian','American'])]
df['citizenship'].value_counts()
```

```python
#DATA VISUALIZATION (ANALYSIS)

#1. Hangi ülkede daha fazla ölüm yaşandı?
sns.set(style="white",palette = "Set1") #grafik özelleştirme..
plt.figure(figsize=(10,6))
plt.rcParams['font.family']='AppleGothic'

# Sayım grafiğini kullanarak ölüm sayısını uyruğa göre görselleştireceğiz..
ax = sns.countplot(x='citizenship', data=df)

#Bu döngü, bir Seaborn bar plot (çubuk grafik) üzerinde her bir 
#çubuğun üzerine değer etiketleri eklemek için 
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.title("Uyruğuna göre toplam ölüm sayisi")
plt.xlabel("Vatandaslik")
plt.ylabel("Ölümler")

plt.xticks(rotation=45)

plt.show()
```

```python
#Filistinli ölümler İsrailli ölümlerden yaklaşık 10 kat daha fazla

#2. En çok can kaybı hangi yılda yaşandı?
df['date_of_death'] = pd.to_datetime(df['date_of_death']) # "date_of_death" sütununu tarih formatına dönüştür

# "date_of_death" sütununa göre yılı çıkar
df['year'] = df['date_of_death'].dt.year

# Yıllara göre ölüm sayısını hesapla
fatality_by_year = df.groupby('year').size().reset_index(name='fatalities')

# En çok ölümün olduğu yılı bul
max_fatalities_year = fatality_by_year[fatality_by_year['fatalities'] == fatality_by_year['fatalities'].max()]

# En çok ölümün olduğu satırı bul
most_fatal_day = fatality_by_year.loc[fatality_by_year['fatalities'].idxmax()]


sns.set(style="whitegrid", palette="Set1")
plt.rcParams['font.family'] = 'AppleGothic'

plt.figure(figsize=(12, 6))
ax = sns.lineplot(x='year', y='fatalities', data=fatality_by_year, markers=True)


# En çok ölümün olduğu yılı gösterir
if not max_fatalities_year.empty:
    max_year = max_fatalities_year.iloc[0]['year']
    max_fatalities = max_fatalities_year.iloc[0]['fatalities']
    ax.annotate(f'Most deaths: {most_fatal_day["year"]}, {max_fatalities} people', (max_year, max_fatalities),
                ha='center', va='bottom', fontsize=12, color='red', xytext=(0, 5),
                textcoords='offset points', arrowprops=dict(arrowstyle='->', color='blue'))

plt.title('Death trends from 2000 to 2023', fontsize=16)
plt.ylim(0, 3000)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of deaths', fontsize=12)
plt.show()
```

```python
#3. 2014'te en çok ölümün yaşandığı ilk 3 bölge hangileri?

df_2014 = df[df['date_of_death'].dt.year == 2014]

# İkamet yerine göre ölüm sayısını hesaplay
residence_death_counts = df_2014['place_of_residence'].value_counts()

#İlk 3 bölgeyi seç
top_3_residences = residence_death_counts.head(3)


sns.set(style="white", palette="Set1")
plt.rcParams['font.family'] = 'AppleGothic'
plt.figure(figsize=(10, 6))

ax = sns.barplot(x=top_3_residences.index, y=top_3_residences.values, palette="Set1")

plt.title('Top 3 Deaths by Place of Residence in 2014', fontsize=16)
plt.xlabel('residence', fontsize=12)
plt.ylabel('fatalities', fontsize=12)

plt.xticks(rotation=45)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

```python
#4. Genel olarak 2000'den 2023'e kadar en çok ölüm hangi bölgede yaşandı?

sns.set(style="white", palette="Set1")
plt.figure(figsize=(10, 6))

sns.barplot(data=df,x=df['event_location'].value_counts().values[:20], y=df['event_location'].value_counts().index[:20])

plt.tight_layout()
plt.show()
```

```python
#5. Ölümlerin meydana geldiği ana alanları görselleştirmek için bir harita kullansaydık ne olurdu?

# Ana bölge koordinatlarını tanımla

district_coords = {
    'Gaza': [31.5, 34.466667],
    'Hebron': [31.532569, 35.095388],
    'Jenin': [32.457336, 35.286865],
    'Nablus': [32.221481, 35.254417],
    'Ramallah': [31.902922, 35.206209],
    'Bethlehem': [31.705791, 35.200657],
    'Tulkarm': [32.308628, 35.028537],
    'Jericho': [31.857163, 35.444362],
    'Rafah': [31.296866, 34.245536],
    'Khan Yunis': [31.346201, 34.306286]
}

# Her ilçedeki ölüm sayısı
district_fatalities = df.groupby('event_location_district').size()

# Ölüm sayısına göre dairenin rengini belirleme işlevi

def get_color(fatalities):
    if fatalities > 500:
        return 'darkred'
    elif fatalities > 100:
        return 'grey'
    elif fatalities > 50:
        return 'orange'
    else:
        return 'green'

# Temel harita oluşturma
m = folium.Map(location=[32, 34.75], zoom_start=8)

for district, coords in district_coords.items():
    fatalities = district_fatalities.get(district, 0)
    folium.Marker(
        location=coords,
        tooltip=f'bölge_adi: {district}, ölü: {fatalities}insanlarin_sayisi',  # Bölge adını ve ölüm sayısını gösterir
        icon=None
    ).add_to(m)
    folium.Circle(
        location=coords,
        radius=np.sqrt(fatalities) * 1200, #yarıçap boyutu 
        color=get_color(fatalities),
        fill=True,
        fill_color=get_color(fatalities),
        fill_opacity=0.6,
    ).add_to(m)

# Katman kontrolü ekle
folium.LayerControl().add_to(m)

m
```

```python
#6. Ölenler arasında ilk 3 yaralanma türü nelerdir?

injury_death_counts = df['type_of_injury'].value_counts()[:3]

plt.figure(figsize=(6, 6))
sns.set(style="white", palette="Set1")

plot = plt.pie(injury_death_counts, labels=['Shooting', 'Explosion', 'Shelling'], autopct='%1.1f%%', startangle=140)

plt.title("Top 3 types of injuries by fatality")
plt.legend(title="Injury Type", labels=["Shooting", "Explosion", "Shelling"], loc="upper right")

plt.show()
```

```python
#7. Cinsiyete ve yaşa göre ölüm sayısı nedir?

plt.figure(figsize=(6, 6))
sns.set(style="white", palette="Set2")

sns.histplot(data=df, x='age', hue='gender',bins=20)

plt.title("Number of deaths by age")
plt.xlabel("Age")
plt.ylabel("Number of deaths")
plt.legend(['Woman', 'Man'])

plt.show()
```
